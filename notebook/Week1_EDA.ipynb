```python
df["name"].sample(30, random_state=42)
```

```python
category_keywords = {
    "Kurta": ["kurta", "kurti"],
    "Dress": ["dress"],
    "Top": ["top", "crop top", "blouse"],
    "Jeans": ["jeans"],
    "Trousers": ["trousers", "pant"],
    "Skirt": ["skirt"],
    "Saree": ["saree"],
    "Lehenga": ["lehenga"],
    "Jacket": ["jacket", "bomber"],
    "Hoodie": ["hoodie"],
    "Shawl": ["shawl"],
    "Shoes": ["shoe", "shoes", "heels", "sandals", "sneakers"],

    "Palazzos": ["palazzo", "palazzos"],
    "Co-Ord Set": ["co-ord", "co ord", "co-ords"],
    "Pullover": ["pullover"],
    "Sweatshirt": ["sweatshirt"],
    "Shorts": ["shorts"],
    "Tunic": ["tunic"],
    "Dupatta": ["dupatta"],
    "Jumpsuit": ["jumpsuit"],
    "Playsuit": ["playsuit"],
    "Cardigan": ["cardigan"],
    "Shrug": ["shrug"],
    "Tights": ["tight", "tights", "leggings"],
}
```

```python
def get_category(name):
    name = str(name).lower().strip()

    for category, keywords in category_keywords.items():
        for keyword in keywords:
            if keyword in name:
                return category

    return "Other"
```

```python
df["category"] = df["name"].apply(get_category)
df["category"].value_counts()
```

```python
def get_gender(row):

    name = str(row["name"]).lower()
    category = str(row["category"]).lower()

    if "women" in name:
        return "Women"

    elif "men" in name:
        return "Men"

    elif "girls" in name:
        return "Girls"

    elif "boys" in name:
        return "Boys"

    elif "unisex" in name:
        return "Unisex"

    elif category in [
        "kurta",
        "lehenga",
        "dupatta",
        "palazzos",
        "skirt",
        "saree",
        "playsuit",
        "jumpsuit",
        "tunic",
        "cardigan",
        "shrug",
        "top",
        "dress",
        "co-ord set",
        "sweatshirt",
        "hoodie",
        "pullover"
    ]:
        return "Women"

    else:
        return "Unknown"


df["gender"] = df.apply(get_gender, axis=1)

df["gender"].value_counts()

```

```python


# Create Tags
df["tags"] = (
    df["name"].fillna("") + " " +
    df["brand"].fillna("") + " " +
    df["category"].fillna("") + " " +
    df["colour"].fillna("") + " " +
    df["description"].fillna("")
)
df[["name","tags"]].head()
# Save Feature Engineered Dataset
df.to_csv("../Data/fashion_feature_engineered.csv", index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df["tags"])
tfidf_matrix.shape
```
